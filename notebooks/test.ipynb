```python
import chromadb
from tqdm import tqdm

client = chromadb.PersistentClient(path="../chroma_db/bao_hiem")

old_collection = client.get_collection("BaoHiem_chunks")
new_collection = client.create_collection("bao_hiem")

batch_size = 1000
total = old_collection.count()

print(f"Total documents: {total}")

for offset in tqdm(range(0, total, batch_size), desc="Copying"):
    batch = old_collection.get(
        include=["documents", "metadatas", "embeddings"],
        limit=batch_size,
        offset=offset
    )

    new_collection.add(
        ids=batch["ids"],
        documents=batch["documents"],
        metadatas=batch["metadatas"],
        embeddings=batch["embeddings"]
    )

print("✅ Done copying!")
```
